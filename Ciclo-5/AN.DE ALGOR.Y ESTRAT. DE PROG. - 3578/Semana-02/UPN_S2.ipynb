```python
def sumarv1(a,b):
  s = a + b #s es una variable local
def sumarv2(a,b):
   global s
   s = a + b #s es una variable global
```

```python
#Probando la funcion
s=77
sumarv1(3,4)
print("s despues de llamar a sumarv1(3,4) es : ", s)
sumarv2(5,3)
print("s despues de llamar a sumarv2(3,4) es : ", s)
```

```python
def Dividir(c,d):
  #Estructura de control
  try:
  #Aqui controlamos lo que queremos
    x=c/d
    return x
  except ZeroDivisionError:
    print("No se puede dividir entre cero")
  except TypeError:
    print("El dato ingresado no es valido")
  finally: #Se ejecuta siempre como parte del try
    print("Fin de la funcion")
```

```python
#Comprobando
c = int(input("Ingrese numero entero: "))
d = int(input("Ingrese un divisor: "))
try:
  Dividir(c,d)
except:
  print("Error")
```

```python
Dividir("a","b")
```

```python
#Aplicando Burbuja
def burbuja(registro, columna):
  n = len(registro)
  for i in range(n-1):
    for j in range(0, n-i-1):
      if registro[j][columna] > registro[j+1][columna]:
        registro[j], registro[j+1] = registro[j+1], registro[j]0
  return registro
```

```python
#Llenando datos
producto = [
    {
        "id": 1,
        "nombre": "Refrigeradora",
        "precio": 1000
    },
    {
        "id": 2,
        "nombre": "Televisor",
        "precio": 500
    },
    {
        "id": 3,
        "nombre": "Laptop",
        "precio": 800
    },
    {
        "id": 4,
        "nombre": "Celular",
        "precio": 300
    }
]
```

```python
#Aplicando burbuja ordenado por el ID
productos_ordenados = burbuja(producto, "id")
print("Ordenando registros por ID")
for reg in productos_ordenados:
  print(reg)
```

```python
#Aplicando burbuja ordenado por el precio
productos_ordenados = burbuja(producto, "precio")
print("Ordenando registros por precio")
for reg in productos_ordenados:
  print(reg)
```
